```python
import pandas as pd
import numpy as np
```

Read data from file

```python
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Data Analysis/Demographic Data Analyzer/adult.data.csv")
data
```

How many of each race are represented in this dataset? This should be a Pandas series with race names as the index labels.

```python
race_count = data['race'].value_counts()
race_count
```

What is the average age of men?


```python
average_age_men = data.loc[ data['sex']=="Male",'age'].mean()
average_age_men
```

```python
average_age = data[["sex", "age"]].groupby("sex").mean()
average_age
```

What is the percentage of people who have a Bachelor's degree?


```python
percentage_bachelors_count = data['education'].value_counts()
percentage_bachelors_count
```

```python
bachelors_count = len( data.loc[ data["education"] == "Bachelors"].value_counts() )
```

```python
education_count = len(data['education'])
```

```python
percentage_bachelors = (bachelors_count / education_count) * 100
print(percentage_bachelors)
```

* What percentage of people with advanced education ("Bachelors", "Masters", or "Doctorate") make more than 50K?
* What percentage of people without advanced education make more than 50K?

```python
array_higher_edu = ["Bachelors", "Masters", "Doctorate" ]
higher_education = data.loc[data["education"].isin(array_higher_edu)] # "IN" use: something.isin(somewhere)
higher_education 
```

```python
array_lower_edu = ["Bachelors", "Masters", "Doctorate" ]
lower_education = data.loc[~data["education"].isin(array_higher_edu)] # "NOT IN": ~something.isin(somewhere)
lower_education 
```

Percentage with salary >50K

```python
higher_education_salary50k_count = len(
    higher_education[higher_education['salary'] == '>50K']
    )
higher_education_rich = (
    higher_education_salary50k_count/len(higher_education)
    ) *100
print(higher_education_rich)
```

```python
lower_education_salary50k_count = len(
    lower_education[lower_education['salary'] == '>50K']
    )
lower_education_rich = (
    lower_education_salary50k_count/len(lower_education)
    ) *100
print(lower_education_rich)
```

```python
lower_education_rich = (len(lower_education.loc[ lower_education['salary'] == '>50K']) / len(higher_education))*100
print(lower_education_rich)
```

What is the minimum number of hours a person works per week (hours-per-week feature)?

```python
min_work_hours = data['hours-per-week'].min()
print(min_work_hours)
```

What percentage of the people who work the minimum number of hours per week have a salary of >50K?

```python
num_min_workers = data.loc[
                           (data['salary'] == ">50K") & 
                           (data["hours-per-week"]<45)
                           ]
num_min_workers 
```

```python
rich_percentage = ( 
    len(num_min_workers) / len(data['hours-per-week'])
    ) *100
print(rich_percentage)
```

What country has the highest percentage of people that earn >50K?

```python
highest_salary_country_count = data.loc[
                                        data['salary'] == '>50K',
                                        'native-country'].value_counts()
highest_salary_country_count
```

```python
country_count = data['native-country'].value_counts()
```

```python
highest_earning_country = (
    (highest_salary_country_count / country_count) * 100).idxmax() # idxmax() -> Return index of first occurrence of maximum over requested axis.
print(highest_earning_country)
```

```python
highest_earning_country_percentage = ((highest_salary_country_count / country_count) * 100).max()
print(highest_earning_country_percentage)
```

Identify the most popular occupation for those who earn >50K in India.

```python
highest_salary_IN = data[
                         (data['salary'] == '>50K') & 
                         (data['native-country'] == 'India')
                         ]
highest_salary_IN
```

```python
top_IN_occupation = highest_salary_IN['occupation'].value_counts()
top_IN_occupation
```

```python
top_IN_occupation = highest_salary_IN['occupation'].value_counts().idxmax()
print(top_IN_occupation)
```
